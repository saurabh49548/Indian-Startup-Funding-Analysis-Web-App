```python
#Total money spend in startup ecosystem
round(df['Amount'].sum())
```

```python
#Maximum amount infused in a startup
df.groupby('StartUp')['Amount'].max().sort_values(ascending=False).head(1).values[0]
```

```python
#Average per startup
df.groupby('StartUp')['Amount'].sum().mean()
```

```python
#Avg per investment
df['Amount'].mean()
```

**Naturall we can say -**

- Startup average > Deal average
(because startups can have multiple funding rounds)
- mean() → How big is a typical deal?
- groupby().sum().mean() → How much does a typical startup raise?

```python
#City wise funding
df.groupby('Location')['Amount'].sum().sort_values(ascending=False).head()
```

```python
#Year wise top Startup
df.groupby(['Year','StartUp'])['Amount'].sum().reset_index().sort_values(by= 'Amount',ascending=False).drop_duplicates('Year').sort_values('Year')
```

```python
#Sector wise analysis
#Top Sectors by Investment Value
df.groupby('Vertical')['Amount'].sum().sort_values(ascending = False).head()
```

```python
#Top Sectors by Startup Count
df.groupby('Vertical')['StartUp'].count().sort_values(ascending = False).head()
```

```python
#Funding Stage Analysis
df.groupby('Round').agg({'Amount' : 'sum','StartUp' : 'count'}).reset_index().sort_values(by= 'Amount',ascending=False)
```

```python
#Year-wise Top Startups
temp = df.groupby(['Year', 'StartUp'])['Amount'].sum().reset_index()
top_startups = temp.sort_values(['Year', 'Amount'],ascending=[True, False]).groupby('Year').head(1).sort_values('Year')
top_startups
```

```python
#Top Investors
df_temp = df.copy()
df_temp['Investors'] = df['Investors'].str.split(',')
top_investor = df_temp['Investors'].explode().str.strip().value_counts()
top_investor
```

```python
top_investor = top_investor[top_investor.index != 'Undisclosed']
top_investor.head(10)
```

```python
#Month vs Year Funding
heatmap_data = df.pivot_table(
    values='Amount',
    index='month',
    columns='Year',
    aggfunc='sum'
)
heatmap_data
```

```python
heatmap_data = heatmap_data.drop(columns=2020) #Because only one month data is available for year 2020
```

```python
heatmap_data = heatmap_data.apply(lambda x : x.fillna(x.mean()))
```
