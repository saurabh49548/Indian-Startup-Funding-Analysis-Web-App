**Change USD $ to INR (Rs)**

```python
def to_inr(dollar):
    inr = dollar * 94.43  #Convert USD to INR
    return inr/10000000   #Converting INR to Carores
```

```python
df['Amount']=df['Amount'].apply(to_inr)
```

```python
df['Amount'].head()
```

```python
#Filling the missing value in the Amount column 
mean= df['Amount'].mean()
median = df['Amount']. median()

print(f'Mean = {mean}')
print(f'Median = {median}')
```

Mean (174.03) is significantly higher than your Median (16.05). This usually means your 'Amount' column has outliers

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds (standard is 1.5 * IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_cleaned = df[(df['Amount'] >= lower_bound) & (df['Amount'] <= upper_bound)]

print(f"Old shape: {df.shape}")
print(f"New shape: {df_cleaned.shape}")
```

- Row count dropped from 3044 to 1782. This happened because the filter (df['Amount'] >= lower_bound) automatically excludes rows where the amount is NaN.

```python
new_mean = df_cleaned['Amount'].mean()
new_median = df_cleaned['Amount'].median()

print(f'New Mean = {new_mean}')
print(f'New Median = {new_median}')
```

Even after removing outliers, your Mean (29.8) is still about 3 times higher than your Median (9.44). This suggests your data is still "right-skewed

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].fillna(new_median)
```

- Using the Median to fill 939 rows (about 30% of your data) is a common practice, but it comes with a trade-off you should consider.
- Filling nearly 1,000 rows with the exact same number (9.44) will create a huge "spike" in your data. This can make your correlations look stronger or weaker than they actually are because you've artificially reduced the variety in the 'Amount' column.

```python
# Final clean-up after filling NaNs
df = df[(df['Amount'] >= lower_bound) & (df['Amount'] <= upper_bound)]
```

```python
#By manual assessment we get the inconsistency in vertical names ----> Resolving that 
#Fix the 'é' (like in Café or Décor)
df['Vertical'] = df['Vertical'].str.replace(r'\\\\xc3\\\\xa9', 'e', regex=True)

# 2. Fix the apostrophe (like in Byju's or Caf's)
df['Vertical'] = df['Vertical'].str.replace(r'\\\\xe2\\\\x80\\\\x99', "'", regex=True)

# 3. Wipe out the broken space junk (\\\\xc2\\\\xa0)
df['Vertical'] = df['Vertical'].str.replace(r'\\\\xc2\\\\xa0', ' ', regex=True)

# 4. Fix Newlines (\\\\Nplatform) found in 'Mobile Growth Hacking'
df['Vertical'] = df['Vertical'].str.replace(r'\\\\n', ' ', regex=True, case=False)

# 5. Final Polish: Fix casing and remove extra whitespace
df['Vertical'] = df['Vertical'].str.strip().str.title()

# 6. Optional: Clean up repetitive suffixes to help grouping
# This merges "Booking Platform" and "Booking App" into just "Booking"
df['Vertical'] = df['Vertical'].str.replace(r'\s+(Platform|App|Marketplace|Service|Solution)s?\b', '', regex=True, case=False)
```

I am using a dual-dataframe approach to handle rows with multiple cities (e.g., "Mumbai / Bangalore").

**1. Why am I doing this?**
*   **To get accurate city counts:** A startup listed as "Mumbai / Bangalore" won't show up in a search for just "Mumbai." By splitting them, I can see the true activity in every city.
*   **To avoid "Money Math" errors:** Using `explode()` duplicates the row. If I'm not careful, a $10M investment will be counted twice ($10M for Mumbai + $10M for Bangalore = $20M). 

**2. The Solution:**
*   **`df` (Original):** I will use this for all **Financial Analysis** (Total/Average Funding) to ensure the dollar amounts are 100% accurate.
*   **`df_city` (Exploded):** I will use this copy for **Geographic Analysis** (City counts and maps) so that every location is represented.


```python
# Create a dedicated copy for geographic analysis
df_by_city = df.copy()
```

```python
#Split and Explode
df_by_city['Location'] = df_by_city['Location'].str.split('/')
df_by_city = df_by_city.explode('Location')
df_by_city['Location'].unique()
```

**Standardize the Location name**

```python
city_mapping = {
    # Spelling Fixes & Standardizing to common names
    'Bengaluru': 'Bangalore',
    'Gurugram': 'Gurgaon',
    'Kolkatta': 'Kolkata',
    'Ahemadabad': 'Ahmedabad',
    'Ahemdabad': 'Ahmedabad',
    'Nw Delhi': 'New Delhi',
    'Delhi': 'New Delhi',
    'Kormangala': 'Bangalore', # Neighborhood in Bangalore
    'Andheri': 'Mumbai',       # Neighborhood in Mumbai
    'Chembur': 'Mumbai',       # Neighborhood in Mumbai
    
    # State names mapped to their primary startup hubs 
    'Haryana': 'Gurgaon',
    'Karnataka': 'Bangalore',
    'Kerala': 'Kochi',
    'Uttar Pradesh': 'Noida',
    
    # Cleaning up common International/Combined labels
    'USA': 'US',
    'SFO': 'San Francisco',
    'NY': 'New York'
}
```

```python
df_by_city['Location'] = df_by_city['Location'].replace(city_mapping)
df_by_city['Location'].unique()
```
